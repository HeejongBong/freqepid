```r
library(tidyverse)
```

```r
library(ebci)
```

```r
results_binom <- read.csv('results_nbinom.csv')
```

```r
results_binom
```

```r
results_ebci = ebci(beta_1 ~ 1, data=results_binom, alpha=0.05,
                    se=se_1, weights=1/se_1^2)
```

```r
results_ebci$df %>%
    select(th_eb, len_eb) %>%
    rename(beta_1 = th_eb) %>%
    mutate(upper = beta_1 + len_eb, lower = beta_1 - len_eb)
```

```r
results_sdw_ebci = ebci(beta_1 ~ 1, data=results_binom, alpha=0.05,
                    se=sdw_1, weights=1/sdw_1^2)
```

```r
results_sdw_ebci$df %>%
    select(th_eb, len_eb) %>%
    rename(beta_1 = th_eb) %>%
    mutate(upper = beta_1 + len_eb, lower = beta_1 - len_eb)
```
